```python
# 2. Import libraries
import nltk
import spacy
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
# 3. Download NLTK resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# 4. Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
# 5. Sample simplified medical text (PubMed-style)
medical_text = """
The patient was diagnosed with diabetes mellitus and hypertension.
She is taking medications regularly and monitoring blood glucose levels.
Early diagnosis and proper treatment improve patient outcomes.
"""
```

```python
nltk.download('punkt_tab')
sentences = sent_tokenize(medical_text)
print("Sentence Tokenization:")
for s in sentences:
    print("-", s)
```

```python
# 7. Word Tokenization (NLTK)
words = word_tokenize(medical_text)
print("\nWord Tokenization:")
print(words)
```

```python
# 8. Stemming using Porter Stemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words if word.isalpha()]
print("\nStemming Output:")
print(stemmed_words)
```

```python
# 9. Lemmatization using NLTK
lemmatizer = WordNetLemmatizer()
lemmatized_words_nltk = [lemmatizer.lemmatize(word) for word in words if word.isalpha()]

print("\nLemmatization Output (NLTK):")
print(lemmatized_words_nltk)
```

```python
# 10. Lemmatization using spaCy (recommended for medical text)
doc = nlp(medical_text)
lemmatized_words_spacy = [token.lemma_ for token in doc if token.is_alpha]

print("\nLemmatization Output (spaCy):")
print(lemmatized_words_spacy)
```

```python
# 11. Comparison of Stemming vs Lemmatization
print("\nComparison (Word | Stem | Lemma):")
for word in ["diagnosed", "medications", "monitoring", "levels"]:
    print(
        word,
        "|",
        stemmer.stem(word),
        "|",
        nlp(word)[0].lemma_
    )
```

```python
# 12. Conclusion note (for discussion section)
print("\nConclusion:")
print(
    "Lemmatization preserves the actual medical meaning of words, "
    "which is crucial in healthcare NLP tasks like diagnosis extraction, "
    "clinical decision support, and medical record analysis."
)
```
